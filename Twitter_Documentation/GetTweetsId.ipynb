```python
def getTweets(corpusFile, targetResultFile):
    import csv
    import time

    counter = 0
    corpus = []

    with open(corpusFile, 'r', newline = '\n') as csvfile:                                                                                          
            lineReader = csv.reader(csvfile, delimiter='\t')
            for row in lineReader:
                corpus.append({"tweet_id": row[0], "label": row[1], "topic": row[2]})

    sleepTime = 2
    trainingDataSet = []

    for tweet in corpus:
        try:
            tweetFetched = api.get_status(tweet["tweet_id"])
            print("Tweet fetched" + tweetFetched.text)
            tweet["text"] = tweetFetched.text
            trainingDataSet.append(tweet)
            time.sleep(sleepTime)

        except:
            print("Inside the exception - no:2")
            continue

    with open(targetResultFile, 'w') as csvfile:
        linewriter = csv.writer(csvfile, delimiter=',', quotechar="\"")
        for tweet in trainingDataSet:
            try:
                linewriter.writerow([tweet["tweet_id"], tweet["text"], tweet["label"], tweet["topic"]])
            except Exception as e:
                print(e)
    return trainingDataSet


sample = "jmir_v23i1e25314_app3.txt"
ResultFile = "ResultFile.csv"
resultFile = getTweets(sample, ResultFile)
```
